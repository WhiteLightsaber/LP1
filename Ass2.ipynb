```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix
```

```python
data = pd.read_csv("dd.csv")
```

```python
columns = data.columns
```

```python
columns = [col_name for col_name in columns if not col_name=="Outcome"]
```

```python
X = data.iloc[:,0:8]
```

```python
X
```

```python
Y = data["Outcome"]
```

```python
Y
```

```python
train_X,test_X,train_Y,test_Y = train_test_split(X,Y,test_size=0.3)
```

```python
train_X
```

```python
train_posmean = train_X[train_Y==1].mean()
train_negmean = train_X[test_Y==0].mean()
train_posstd = train_X[test_Y==1].std()
train_negstd = train_X[test_Y==0].std()
```

```python
summary = {"POS MEAN":train_posmean.tolist(),"NEG MEAN":train_negmean.tolist(),"POS STD":train_posstd.tolist(),"NEG STD":train_negstd.tolist()}
```

```python
summary
```

```python
def gaussian(x,mn,std):
    var = std*std
    p = 1/(np.sqrt(2*np.pi*var))*np.exp(-(x-mn)**2/(2*var))
    print(p)
    return p
```

```python
def predict(row, summary):
    prob_pos = len(summary["POS MEAN"])/(len(summary["POS MEAN"])+len(summary["NEG MEAN"]))
    neg_pos = len(summary["NEG MEAN"])/(len(summary["POS MEAN"])+len(summary["NEG MEAN"]))
    for i in range(0,len(row)):
        prob_pos = prob_pos*gaussian(row[i],summary["POS MEAN"][i],summary["POS STD"][i])
    for i in range(0,len(row)):
        neg_pos = neg_pos*gaussian(row[i],summary["NEG MEAN"][i],summary["NEG STD"][i])
    return [prob_pos,neg_pos]
```

```python
predictions_raw = []
for row in test_X.values.tolist():
    predictions_raw.append(predict(row, summary))
```

```python
predictions_raw
```

```python
predictions = []
for row in predictions_raw:
    if(row[0]>row[1]):
        predictions.append(1)
    else:
        predictions.append(0)
```

```python
confusion_matrix(test_Y.tolist(),predictions)
```
