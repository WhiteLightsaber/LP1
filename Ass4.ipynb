```python
df=df.drop('Start date',axis=1)
df=df.drop('End date',axis=1)
df=df.drop('Start station',axis=1)
df=df.drop('End station',axis=1)
```

```python
df
```

```python
le = LabelEncoder()
```

```python
le.fit(df["Member type"])
```

```python
df["Member type"] = le.transform(df["Member type"])
```

```python
df
```

```python
le = LabelEncoder();
le.fit(df["Bike number"])
df["Bike number"]=le.transform(df["Bike number"])
```

```python
df
```

```python
train = np.array(df.iloc[0:10000])
```

```python
test = np.array(df.iloc[100000:,])
```

```python
test
```

```python
train
```

```python
from sklearn.naive_bayes import GaussianNB
```

```python
model = GaussianNB()
```

```python
model.fit(train[:,0:4],train[:,4])
```

```python
predicted = model.predict(test[:,0:4])
```

```python
predicted
```

```python
count=0                 ### accuracy
for l in range(30597):  
    if(predicted[l]==test[l,4]):
        count=count+1
```
